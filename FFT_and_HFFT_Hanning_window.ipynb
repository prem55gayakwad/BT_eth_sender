```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.signal import hanning
```

```python
# Read data from file 'filename.csv' 
data = pd.read_csv("blob6_ut.csv") 

#amp=data['amp']
#time= data['t']
time= data['t']
amp=data['amp']
```

```python
# x axis values
time1 = time
# corresponding y axis values
amp1 = amp
# plotting the points 
plt.plot(time1, amp1)
# naming the x axis
plt.xlabel('time (s)')
# naming the y axis
plt.ylabel('amplitude')
# giving a title to my graph
plt.title('Orginal signal')
# function to show the plot
plt.show()
```

```python
# Number of sample points
N = len(amp1)

#Calculate fft
yf = np.fft.fft(amp1)

#Calculate fft with Hanning window
w = hanning(len(yf))
ywf = np.fft.fft(amp1*w)

#To find absolute value and FFT is symmetric, throw away second half
abs_yf=np.abs(yf[1:N//2])
abs_ywf=np.abs(yf[1:N//2])

#To find length of FFT
len_yf=len(np.abs(yf[1:N//2]))
len_ywf=len(np.abs(ywf[1:N//2]))

#----------------------------------------------------------------
#Calculate HFFT
yhf = np.fft.hfft(amp1)

#Calculate Hfft with Hanning window
wh = hanning(len(yhf))
hfft_hann=yhf*wh
ywhf = np.fft.fft(hfft_hann)

#To find absolute value and HFFT is symmetric, throw away second half
abs_yhf=np.abs(yhf[1:N//2])
abs_ywhf=np.abs(yhf[1:N//2])

#To find length of HFFT
len_yhf=len(np.abs(yhf[1:N//2]))
len_ywhf=len(np.abs(ywhf[1:N//2]))
```

```python
#Plot FFT signal
uz=(time[0:len_yf])
plt.plot(time[0:len_yf], np.abs(yf[1:N//2]))
# naming the x axis
plt.xlabel('Frquency (Hz)')
# naming the y axis
plt.ylabel('amplitude')
# giving a title to my graph
plt.title('FFT signal')
plt.show()
```

```python
#Plot FFT signal with Hanning Window
plt.plot(time[0:len_ywf], np.abs(ywf[1:N//2]))
# naming the x axis
plt.xlabel('Frquency (Hz)')
# naming the y axis
plt.ylabel('amplitude')
# giving a title to my graph
plt.title('FFT signal with Hanning window')
plt.show()
np.abs(yf[1:N//2])
```
